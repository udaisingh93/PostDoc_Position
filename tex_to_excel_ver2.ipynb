```python
import xlsxwriter
import re
import numpy as np
import glob
import math
```

```python
def text_to_excel(ang,sigma,txt):
    print("newcross10Mev/doublediff_cossat{1}_*.txt".format(txt,ang,sigma))
    mylist = [f for f in glob.glob("./Tritonnewcross/{0}doublediff_cossat{1}_*.txt".format(txt,ang,sigma))]
    print(mylist)
    
    workbook = xlsxwriter.Workbook('{0}Proton{1}.xlsx'.format(txt,ang,sigma))
    worksheet = workbook.add_worksheet("Proton")
    worksheet1 = workbook.add_worksheet("Deuteron")
    worksheet2 = workbook.add_worksheet("Triton")
    worksheet3 = workbook.add_worksheet("Pion")
    row=2
    col=0
    i=1
    for file in mylist:
        row=2
        sig =re.split("_",file)
        #print(sig[2])
        f = open(file, 'r')
        for line in f:
        #print(line)
            if(re.match(r"Proton",line)):
                worksheet.write('A1', line)
            elif(re.match(r"Deuteron",line)):
                break
            else:
                sp =re.split(" |\n",line)
                flag2= not math.isnan(float(sp[3]))
                #print(sp)
                if(col==0and flag2):
                    worksheet.write(1, col,"Energy" )
                    worksheet.write(row, col,     sp[0])
                if(flag2):
                    worksheet.write(1, col+1,f"Cross_sec{sig[2]}" )
                    worksheet.write(1, col+2,f"StatErr{sig[2]}" )
                    worksheet.write(1, col+3,f"EffErr{sig[2]}" )
                    worksheet.write(row, col + 1, sp[1])
                    worksheet.write(row, col + 2, sp[2])
                    worksheet.write(row, col + 3, sp[3])
                    row += 1
        row=2
        for line in f:
            worksheet1.write('A1', "Deuteron")
            if(re.match(r"triton",line)):
                break
            else:
                sp =re.split(" |\n",line)
                flag2= not math.isnan(float(sp[2]))
                #print(sp)
                if(col==0and flag2): 
                    worksheet1.write(1, col,"Energy" )
                    worksheet1.write(row, col,     sp[0])
                if(flag2):
                    worksheet1.write(1, col+1,f"Cross_sec{sig[2]}" )
                    worksheet1.write(1, col+2,f"StatErr{sig[2]}" )
                    worksheet1.write(1, col+3,f"EffErr{sig[2]}" )
                    worksheet1.write(row, col + 1, sp[1])
                    worksheet1.write(row, col + 2, sp[2])
                    worksheet1.write(row, col + 3, sp[3])
                    row += 1
        row=2
        for line in f:
            worksheet2.write('A1', "Triton")
            if(re.match(r"pion",line)):
                break
            else:
                sp =re.split(" |\n",line)
                flag2= not math.isnan(float(sp[2]))
                #print(sp)
                if(col==0and flag2): 
                    worksheet2.write(1, col,"Energy" )
                    worksheet2.write(row, col,     sp[0])
                if(flag2):
                    worksheet2.write(1, col+1,f"Cross_sec{sig[2]}" )
                    worksheet2.write(1, col+2,f"StatErr{sig[2]}" )
                    worksheet2.write(1, col+3,f"EffErr{sig[2]}" )
                    worksheet2.write(row, col + 1, sp[1])
                    worksheet2.write(row, col + 2, sp[2])
                    worksheet2.write(row, col + 3, sp[3])
                    row += 1
        row=2
        for line in f:
            worksheet3.write('A1', "Pion")
            if(re.match(r"triton",line)):
                break
            else:
                sp =re.split(" |\n",line)
                flag2= not math.isnan(float(sp[3]))
                #print(sp)
                if(col==0and flag2): 
                    worksheet3.write(1, col,"Energy" )
                    worksheet3.write(row, col,     sp[0])
                if(flag2):
                    worksheet3.write(1, col+1,f"Cross_sec{sig[2]}" )
                    worksheet3.write(1, col+2,f"StatErr{sig[2]}" )
                    worksheet3.write(1, col+3,f"EffErr{sig[2]}" )
                    worksheet3.write(row, col + 1, sp[1])
                    worksheet3.write(row, col + 2, sp[2])
                    worksheet3.write(row, col + 3, sp[3])
                    row += 1
        col+=3
        i=i+1
    workbook.close()
```

```python
INCL=""
sigma=0.6
for ang in [20,25,30,35,40,45,50,55,60,65,70,75,80]:
    text_to_excel(ang,sigma,INCL)
```

```python
import pandas as pd
import numpy as np
import re
uplimit=[[3000,3150,3150,3150,3150,3150,3150,3150,3000,3010,3000,3000,3000,3000,3000],[850,850,820,820,820,820,820,820,850,450,350],[400,400,400,400,400,400,400,400,450,450,450,450,450,450],[800,800,800,800,800,800,600,600,800,850,850,850]]
Particles = [
  "Proton",
  "Deuteron",
  "Triton",
  "PionPositive"
]
pd.set_option('precision', 4)
def read_excel(file,sheet,ang):
    print(file)
    upl=uplimit[sheet]
    if ang==25:
        up=upl[0]
    elif ang==35:
        up=upl[1]
    elif ang==65:
        up=upl[2]
    else:
        up=upl[3]
    particle=Particles[sheet]
    sp =re.split("\.",file)
    print(sp[0])
    df=pd.read_excel(file,skiprows=1,sheet_name=sheet)
    headers=["Energy","Sector1","Error 1","Sector2","Error 2","Sector3","Error 3","Sector4","Error 4","Sector5","Error 5","Sector6","Error 6"]
    df2=pd.read_excel(f"../crossparticle/{sp[0]}_1.0.xlsx",header=None,skiprows=1,sheet_name=sheet)
    df2=df2.loc[:,0:12]
    df2.columns=headers
    
    df=df[df['Energy']>40]
    df2=df2[df2['Energy']>40]
    df=df[df['Energy']<up]
    df2=df2[df2['Energy']<up]
    df2["Mean"]=df2.iloc[:,1::2].mean(axis=1)
    df = df.replace(0, np.nan)
    df2=df2.dropna()
    if(sheet==2 or sheet==3 or sheet==1):
        df=df.iloc[:,0:9]
    df.dropna(inplace=True)
    df2.dropna(inplace=True)
    df.reset_index(drop=True, inplace=True)
    df2.reset_index(drop=True, inplace=True)
    df2["Abs"]=np.abs(df2.iloc[:,1]-df2['Mean'])
    for i in range(3,12,2):
        df2["Abs"]+=np.abs(df2.iloc[:,i]-df2['Mean'])
    df2["Abs"]=df2["Abs"]/12
    df2["Per"]=(df2["Abs"]/df2['Mean'])*100
    df['Mean']=df.iloc[:,1::3].mean(axis=1)
    df['PIDerr']=df.iloc[:,1::3].std(axis=1)
    df3=df[["Energy","Mean","PIDerr","StatErr0.6","EffErr0.6"]]
    df3=pd.concat([df3, df2['Abs']], axis=1, sort=False)
    df3.dropna(inplace=True)
    df3.rename(columns = {'Mean':'Cross Section','Abs':'SectorErr','StatErr0.6':'StatErr','EffErr0.6':'EffError'}, inplace = True)
    df3["SysErr"]=np.sqrt(df3.iloc[:,2]**2+df3.iloc[:,3]**2+df3.iloc[:,4]**2)
    df4=df3[['Energy','Cross Section','SysErr','StatErr']]
    df4.drop(df4[df4['Cross Section'] < df4['SysErr']].index, inplace = True) 
    formats = {'Energy': '{:3.1f}', 'Cross Section': '{:.3E}', 'SysErr': '{:.3E}','StatErr': '{:.3E}'}
    for col, f in formats.items():
        df4[col] = df4[col].map(lambda x: f.format(x))
#     if(sheet==1 or sheet==2):
#         df4=df4[::4]
    df4.to_csv(f'{particle}_{ang}.txt', index = False, header=True,sep="\t")
    df3.to_csv(f'{particle}_{ang}piderror.csv', index = False, header=True)
    return df,df3,df4
```

```python
for ang in [20,25,30,35,40,45,50,55,60,65,70,75,80]:
#for ang in [20,25,55,65,80]:
    df,df2,df3=read_excel(f"Proton{ang}.xlsx",2,ang)
```
